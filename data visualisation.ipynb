```python
random_ = np.loadtxt("./Filon/random.txt")
green_ = np.loadtxt("./Filon/green.txt")
for i in ["Bessel","series","asy"]:
    input_ = np.loadtxt("./"+i+"/random.txt")
    output_ = np.loadtxt("./"+i+"/green.txt")
    random_ = np.append(random_,input_,axis =0)
    green_ =np.append(green_,output_,axis = 0)
```

```python
input_ = random_ 
output_ = green_
```

```python
input_.shape,output_.shape
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_x = pd.DataFrame(columns=["X","Y","Z1","Z2","Z3"])
```

```python
# df =df_x.sample(frac=0.5)
```

```python
df_x["X"] ,df_x["Y"],df_x["Z1"],df_x["Z2"],df_x["Z3"] = input_[:,0],input_[:,1],output_[:,0],output_[:,1],output_[:,2]
```

```python
df = df_x[df_x['Y'] <= 10]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(df['X'], df['Y'], df['Z1'], cmap=plt.cm.viridis, linewidth=0.2)
surf=ax.plot_trisurf(df['X'], df['Y'], df['Z1'], cmap=plt.cm.viridis, linewidth=0.2)
fig.colorbar( surf, shrink=0.5, aspect=5)
ax.view_init(30, 45)
ax.plot_trisurf(df['X'], df['Y'], df['Z1'], cmap=plt.cm.jet, linewidth=0.01)
ax.set_ylabel("beta")
ax.set_xlabel('tau')
ax.set_zlabel("G",rotation=30)
plt.savefig("plot.eps", format='eps')
plt.show()
```
